Importing the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Quiver plot

```python
# Creating the grids

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)

X , Y = np.meshgrid(x,y)

# Initializing the velocity components
u = -1 - X**2 + Y
v = 1 + X - Y**2

# calculating the velocity magnitude
vel_mag = np.sqrt(u**2 + v**2)


#plotting

plt.figure(figsize=(10,10))
plt.streamplot(X,Y,u,v,color=vel_mag)
plt.colorbar(label = 'VELOCITY MAGNITUDE')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.title('BASIC_STREAM_LINED_PLOT')
plt.show()
```

1 - Dimensional Steady State Heat Conduction Equation

```python
# Required Parameters
L = 1.0
p = 10
dx = L/ (p-1)
k = 100
T_left = 100
T_right = 25

# Matrix and Vector Initialisation

A = np.zeros((p,p))
b = np.zeros(p)


for i in range(1,(p-1)):
    A[i , i-1] = 1
    A[i , i] = -2
    A[i , i+1] = 1
A[0,0] = 1
A[-1 ,-1] = 1

b[0] = T_left
b[-1] = T_right

# solving the system
T = np.linalg.solve(A,b)

# plotting the results

x = np.linspace(0, L, p)
plt.plot(x, T, label = 'T')
plt.xlabel('Rod Position in Metre')
plt.ylabel('Temperature (Celscius)')
plt.title('1-Dimensional Steady State Heat Conduction')
plt.legend()
plt.show()

```

Contour plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(-5, 5, 100)  # X-axis values
y = np.linspace(-5, 5, 100)  # Y-axis values
X, Y = np.meshgrid(x, y)  # Create a grid of X and Y values

# Define a function for Z values
Z = np.sin(np.sqrt(X**2 + Y**2))  # Example: radial sine wave

# Create contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')  # Add contour lines
plt.clabel(contour, inline=True, fontsize=8)  # Add labels to contour lines
plt.colorbar(contour, label='Z value')  # Add color bar

# Add titles and labels
plt.title("Contour Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display the plot
plt.show()
```
